```python
# First Import the Pandas, pathlib, & Numpy Packages

import pandas as pd
import numpy as np
import pathlib
```

```python
# Set Up Export and Export Paths
root_path = r'C:\Users\MLeif\data_science_practice'
project_folder = '1_descriptive_stats'
data_folder = 'datasets'

import_path = pathlib.Path(root_path).joinpath(project_folder).joinpath(data_folder)
export_path = pathlib.Path(root_path).joinpath(project_folder).joinpath(data_folder)

print(import_path)
print(export_path)
```

```python
#Import the cleaned Film Dataset
import_data_file = 'imdb_film_clean.csv'
import_file = import_path.joinpath(import_data_file)

imdb_movies = pd.read_csv(import_file).drop_duplicates()
imdb_movies['movie_id'] = imdb_movies.index
imdb_movies = imdb_movies.set_index('movie_id')
imdb_movies.head()
```

```python
#Question 1: What are the top 20 Films with the Most Votes

top_20_movies = imdb_movies.sort_values('total_votes',ascending=False).head(20)

print(str(top_20_movies[['film_name','total_votes']]))
```

```python
# Create a subset of columns with only Genre Columns

genere_list = ['genre_action','genre_adventure','genre_western',
       'genre_animation','genre_biography','genre_comedy','genre_crime',
       'genre_documentary','genre_drama','genre_family','genre_fantasy',
       'genre_film-noir','genre_history','genre_horror','genre_music',
       'genre_musical','genre_mystery','genre_romance','genre_sci-fi',
       'genre_short','genre_sport','genre_thriller','genre_war']

genre_col = ['movie_id'] + [genere_list]

movie_genre = imdb_movies[genere_list]
```

```python
# For the Next Few Questions Create a Dataset of Just Genre Flags

genre_melt = movie_genre.reset_index().melt(id_vars=['movie_id'], value_vars=[
       'genre_action','genre_adventure','genre_western',
       'genre_animation','genre_biography','genre_comedy','genre_crime',
       'genre_documentary','genre_drama','genre_family','genre_fantasy',
       'genre_film-noir','genre_history','genre_horror','genre_music',
       'genre_musical','genre_mystery','genre_romance','genre_sci-fi',
       'genre_short','genre_sport','genre_thriller','genre_war'
       ]).\
       set_index('movie_id').\
       rename(columns = {'variable':'genre'}).\
       sort_values('genre', ascending=True)

genre_melt = genre_melt[genre_melt['value'] == 1].drop('value', axis=1)
genre_melt['genre'] = genre_melt['genre'].str.replace('genre_','')

genre_melt.head()
```

```python
#Question 2: How Many Films are in Each Genre

movie_genre_totals = pd.DataFrame(genre_melt['genre'].value_counts()).reset_index().\
    rename(columns = {'index':'genre','genre':'Total_Films'})

print(movie_genre_totals)
```

```python
# Question 3: What are the Top 5 and Bottom 5 Film Genres

top_5_genres = movie_genre_totals['genre'].head(5)
bottom_5_genres = movie_genre_totals['genre'].tail(5)

print('Top 5 Film Genres: \n' + str(top_5_genres) + '\n')
print('Bottom 5 Film Genres: \n' + str(bottom_5_genres))
```

```python
# Question 4: What are the top 10 Different Genre Combinations

genre_string = genre_melt.reset_index().\
    groupby(['movie_id'])['genre'].\
    apply(lambda x: '_'.join(x)).reset_index()

multi_genre = genre_string[genre_string['genre'].str.contains('_')]

top_genre_combos = pd.DataFrame(multi_genre.reset_index().\
    groupby('genre')['movie_id'].count()).\
    sort_values('movie_id', ascending = False).\
    rename(columns = {'movie_id':'total_movies'}).iloc[0:10]

print(top_genre_combos)
```

```python
# What is the Median and Percentile Number of Vores for each Film

votes_by_genre = imdb_movies[['total_votes']].join(genre_melt)

genre_votes_summary = votes_by_genre.groupby('genre')['total_votes'].\
    agg([np.mean, np.std, np.median])

genre_votes_summary = genre_votes_summary.join(votes_by_genre.groupby('genre').quantile(.10).\
    sort_values('total_votes', ascending=False).\
    rename(columns={'total_votes':'10_pct'}))

genre_votes_summary = genre_votes_summary.join(votes_by_genre.groupby('genre').quantile(.25).\
    sort_values('total_votes', ascending=False).\
    rename(columns={'total_votes':'25_pct'}))

genre_votes_summary = genre_votes_summary.join(votes_by_genre.groupby('genre').quantile(.75).\
    sort_values('total_votes', ascending=False).\
    rename(columns={'total_votes':'75_pct'}))

genre_votes_summary = genre_votes_summary.join(votes_by_genre.groupby('genre').quantile(.90).\
    sort_values('total_votes', ascending=False).\
    rename(columns={'total_votes':'90_pct'}))

print('Top 5 Genres with the Highest Mean Votes: ' + str(genre_votes_summary.sort_values('mean')['mean'].tail(5)))
print('Top 5 Genres with the Highest Median Votes: ' + str(genre_votes_summary.sort_values('median')['median'].tail(5)))
print('Top 5 Genres with the Highest Stdev of votes: ' + str(genre_votes_summary.sort_values('std')['std'].tail(5)))
print('Top 5 Genres with the Highest Mean Votes: ' + str(genre_votes_summary.sort_values('mean')['mean'].head(5)))
print('Top 5 Genres with the Highest Median Votes: ' + str(genre_votes_summary.sort_values('median')['median'].head(5)))
print('Top 5 Genres with the Highest Stdev of votes: ' + str(genre_votes_summary.sort_values('std')['std'].head(5)))
```

```python
duration_by_genre = imdb_movies[['film_duration']].join(genre_melt)

genre_duration_summary = duration_by_genre.groupby('genre')['film_duration'].\
    agg([np.mean, np.std, np.median])

genre_duration_summary = genre_duration_summary.join(duration_by_genre.groupby('genre').quantile(.10).\
    sort_values('film_duration', ascending=False).\
    rename(columns={'film_duration':'10_pct'}))

genre_duration_summary = genre_duration_summary.join(duration_by_genre.groupby('genre').quantile(.25).\
    sort_values('film_duration', ascending=False).\
    rename(columns={'film_duration':'25_pct'}))

genre_duration_summary = genre_duration_summary.join(duration_by_genre.groupby('genre').quantile(.75).\
    sort_values('film_duration', ascending=False).\
    rename(columns={'film_duration':'75_pct'}))

genre_duration_summary = genre_duration_summary.join(duration_by_genre.groupby('genre').quantile(.90).\
    sort_values('film_duration', ascending=False).\
    rename(columns={'film_duration':'90_pct'}))

print('Top 5 Genres with the Highest Mean duration: ' + str(genre_duration_summary.sort_values('mean')['mean'].tail(5)))
print('Top 5 Genres with the Shortest Mean duration: ' + str(genre_duration_summary.sort_values('mean')['mean'].head(5)))
print('Top 5 Genres with the Highest Median duration: ' + str(genre_duration_summary.sort_values('median')['median'].tail(5)))
print('Top 5 Genres with the Shortest Median duration: ' + str(genre_duration_summary.sort_values('median')['median'].head(5)))
print('Top 5 Genres with the Highest Stdev of duration: ' + str(genre_duration_summary.sort_values('std')['std'].tail(5)))
print('Top 5 Genres with the Lowest Stdev of duration: ' + str(genre_duration_summary.sort_values('std')['std'].head(5)))
```
